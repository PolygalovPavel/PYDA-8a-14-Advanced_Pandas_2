### Задание 1

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings.head()
```

```python
ratings_user = ratings.groupby('userId').agg('count').reset_index()[['userId', 'rating']]
ratings_user.rename(columns={'rating':'count_rating'}, inplace=True)
ratings_user = ratings_user.loc[lambda row: row['count_rating'] > 100]
ratings_user.head()
```

```python
filtr_ratings = ratings_user.join(ratings.set_index('userId'), on='userId', how='left')
filtr_ratings.head()
```

```python
ratings_grouped = filtr_ratings.groupby('userId').agg([min, max])
ratings_grouped.head()
```

```python
ratings_grouped['diff'] = ratings_grouped['timestamp']['max'] - ratings_grouped['timestamp']['min']
ratings_grouped.head()
```

```python
ratings_grouped['diff'].mean() / 24 / 3600
```
